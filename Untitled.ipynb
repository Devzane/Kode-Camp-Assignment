```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Reading Image imread(Takes path of the image as input and return a numpy array)
img1 = cv2.imread(r"C:\Users\Zeetha Robotics\Documents\my properties\download\DSC_0003.JPG")
img2 = cv2.imread(r"C:\Users\Zeetha Robotics\Pictures\Screenshots\Screenshot (2).png")

# Writing images imwrite(Takes path image and image data as input)

img1_write = cv2.imwrite(r"C:\Users\Zeetha Robotics\Documents\img1_write.png", img1)
img2_write = cv2.imwrite(r"C:\Users\Zeetha Robotics\Pictures\img2_write.png", img2)

#showing the images imshow(takes the windowname and the image data as input)
plt.imshow(img1)
plt.axis()
plt.show()

plt.imshow(img2)
plt.axis()
plt.show()
```

```python
#converting Color In images Using openCv

#To convert from BGR to RGB, we use the cvtColor(images, conversion)

img_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2_rgb = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis('off')
plt.show()


plt.imshow(img2_rgb)
plt.axis('off')
plt.show()
```
