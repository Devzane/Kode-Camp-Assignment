```python
#Task 2.1
img_color = cv.cvtColor(img_color, cv.COLOR_BGR2RGB)
# Function for image resizing

def resize_img(img, size):
    resized_img = cv.imresize(img, size)
    return resized_img

# Function for cropping images

def crop_img(img, x1, y1, x2, y2):
    cropped_img = img[y1:y2, x1:x2]

    return cropped_img

# function to rotate image

def rotate_image(img, angle):
    h, w = img.shape[:2]
    center = (w // 2, h // 2)
    matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(img, matrix, (w, h))
    return rotated

```

```python
# Resizing to smaller version
resized = resize_img(img_color, (150, 150))

# Croping a portion
cropped = crop_img(img_color, 100, 100, 300, 300)

# Rotating by 45 degrees
rotated = rotate_img(img_color, 45)

# Displaying all
fig, axes = plt.subplots(1, 4, figsize=(20,5))
axes[0].imshow(img_color)
axes[0].set_title('Original')
axes[1].imshow(resized)
axes[1].set_title('Resized')
axes[2].imshow(cropped)
axes[2].set_title('Cropped')
axes[3].imshow(rotated)
axes[3].set_title('Rotated')

for ax in axes:
    ax.axis('off')
plt.show()
```
